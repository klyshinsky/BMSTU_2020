Для начала нарисуем простую кривую: $y=x^2$.

Вот <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot">здесь</a> описаны все варианты задания цветов, точек и формы штриха.
Заодно добавим сетку - plt.grid().

```python
x=[i/20 for i in range(0,40)]
y=[i*i for i in x]

plt.plot(x, y, 'b--')
plt.plot(x, y[::-1], color='#FACE8D', marker='^')

plt.grid()

_=plt.show()
```

А еще добавим легенду. Для этого сперва сохраним наши графики в отдельные переменные plot1 и plot2.
После этого вызовем функцию plt.legend, в которую передается график (раз мы хотим его показать на легенде) и подпись к нему. loc = 'best' означает, что мы просим библиотеку найти лучшее с ее точки зрения размещение легенды.
Всё о легенде <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html#matplotlib.pyplot.legend">здесь</a>.

```python
x=[i/20 for i in range(0,40)]
y=[i*i for i in x]

plot1, =plt.plot(x, y, 'b--')
plot2, =plt.plot(x, y[::-1], color='#FACE8D', marker='^')

plt.grid()
plt.legend((plot1, plot2), ("График 1", "График 2"), loc = 'center')

_=plt.show()
```

Всё то же самое можно было бы получить, добавив метку (label) к каждому графику, а потом попросив нарисовать легенду.<br>
Заодно окрасим легенду в другой цвет (<font color='#448844'>#448844</font>), увеличим размер шрифта (fontsize=12), добавим название графика ( plt.title() ) и название осей (xlabel() и ylabel()). В итоге легенда перебегает на другое место.

```python
x=[i/20 for i in range(0,40)]
y=[i*i for i in x]

plot1, =plt.plot(x, y, 'b--', label="График 1")
plot2, =plt.plot(x, y[::-1], color='#FACE8D', marker='^', label="График 2")

plt.grid()
plt.title("Здесь должно быть название")
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
legend=plt.legend(loc = 'best', fontsize=12)
legend.get_frame().set_facecolor('#448844')


_=plt.show()
```

А еще можно сделать два графика на одном рисунке и подпись к графику, включающую текст на TeX'е.
Для построения нескольких графиков необходимо создать subplot, в который мы будем рисовать до создания следующего subplot или отображения рисунка.<br>
В график можно вставлять аннотации при помощи plt.annotate(). В ту функцию передается текст, кооринаты аннотируемой точки (xy), координаты текста аннотации (xytext) и стрелочка (arrowprops).<br>
plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),<br>
            arrowprops=dict(facecolor='black', shrink=0.05),)<br>
Помимо этого на графике можно размещать произвольный текст при помощи plt.text(). Весь текст, который мы выводим может содержать в себе формулы, записанные в Tex'е. Для этого необходимо поставить их в знаки доллара и написать формулу. Например так: \$y=x^2\$ даст нам $y=x^2$.

```python
plt.subplot(221)
x=[i/20 for i in range(0,40)]
y=[i*i for i in x]
plot1, =plt.plot(x, y, 'b--', label="$y=x^2$")
plot2, =plt.plot(x, y[::-1], color='#FACE8D', marker='^', label="$y=(x-2)^2$")

plt.grid()
legend=plt.legend(loc = 'best', fontsize=12)
legend.get_frame().set_facecolor('#448844')
plt.annotate('равны $x^2$', xy=(1, 0.9), xytext=(0.9, -0.1),
            arrowprops=dict(facecolor='black', shrink=0.05),)


ax=plt.subplot(224)
x=[i/20 for i in range(0,40)]
y=[math.sin(i) for i in x]
ax.plot(x, y, 'k*')
ax.text(0.9, 0.6, r'$\mathcal{A}\mathrm{sin}(\omega t)$',fontsize=20)

plt.subplot(223)
x=[i/20 for i in range(0,40)]
y=[math.sin(i) for i in x]
plt.plot(x, y, 'k*')
plt.text(0.9, 0.6, r'$\mathcal{A}\mathrm{sin}(\omega t)$',fontsize=20)

_=plt.show()
```

А теперь давайте попробуем рисовать при помощи функции scatter, которая работает несколько медленнее, но обладает большей функциональностью (см. <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter">здесь</a>).<br>
Сперва попробуем сделать всё то же самое. Так как scatter по определению - это набор точек, которые не должны соединяться между собой, а просто бросаются на плоскость, то отрисовать линию с его помощью не получится. Оставляем plot(). <br>
Также обратите внимание, что в scatter нельзя передать сразу цвет и вид точки/штриха, поэтому используем параметры color и marker.

```python
plt.subplot(211)
x=[i/20 for i in range(0,40)]
y=[i*i for i in x]
plot1, =plt.plot(x, y, 'b--')
plot2=plt.scatter(x, y[::-1], color='#FACE8D', marker='^')

plt.grid()
legend=plt.legend((plot1, plot2), ("$y=x^2$", "$y=(x-2)^2$"), loc = 'best', fontsize=12)
legend.get_frame().set_facecolor('#448844')
plt.annotate('равны', xy=(1, 0.9), xytext=(0.9, -0.1),
            arrowprops=dict(facecolor='black', shrink=0.05),)


plt.subplot(212)
x=[i/20 for i in range(0,40)]
y=[math.sin(i) for i in x]
plt.scatter(x, y, color='k', marker='*')
plt.text(0.9, 0.6, r'$\mathcal{A}\mathrm{sin}(\omega t)$',fontsize=20)

_=plt.show()
```

Теперь попробуем особенности scatter. <br>
Параметр alpha можно применять почти везде. Он показывает степень прозрачности графика, легенды, ... 0 - полностью прозрачен, 1 - абсолютно непрозрачен. <br>
Помимо этого для каждой из точек можно задавать размер маркера (параметр s) и его цвет (параметр c).

```python
x=[i/100 for i in range(0,100)]
y=[i*i for i in x]
c=[i*i for i in x]
s=[i*i*100 for i in x]

plt.scatter(x, y, c=c, s=s, alpha=0.6)

_=plt.show()
```

Цвет графика можно задать при помощи цветовой карты. Для использования стандартных цветовых карт необходимо загрузить их из matplotlib: <br>
from matplotlib import cm <br>
А потом использовать ту, что больше нравится, в параметре cmap (а можно создать объект типа Colormap самостоятельно и передать туда же). В этом случае параметр <b>c</b> воспринимается как номер цвета в цветовой схеме, цветовая схема распределяется между минимальным и максимальным значениями. (И никто не мешал сделать то же самое с plot).

```python
from matplotlib import cm


x=[i/100 for i in range(0,100)]
y=[i*i for i in x]
c=[i*i for i in x]

plt.scatter(x, y, c=c, alpha=0.5, cmap=cm.coolwarm)
plt.scatter([x1*2 for x1 in x], y, c=c, alpha=0.5, cmap=cm.autumn)

_=plt.show()
```

А теперь сделаем следующий фокус. Сгенерируем матрицу точек, цвет точки будет задавать значение функции от двух переменных.<br>
Вопрос: какова размерность полученного графика?

```python
x1=[]
y1=[]
z1=[]
for r in range(20):
    for r2 in range(20):
        x1.append(r)
        y1.append(r2)
        z1.append(2*r+r2)

plt.scatter(x1, y1, c=z1, alpha=0.5, cmap=cm.jet)

_=plt.show()
```

Правильно, график трехмерный: две координаты и значение функции.<br>
И, да, я знаю, что с помощью numpy это делается проще.<br>
Теперь добавим размер точки, чтобы отобразить четверую координату.<br>

```python
x1=[]
y1=[]
z1=[]
s1=[]
for r in range(20):
    for r2 in range(20):
        x1.append(r)
        y1.append(r2)
        z1.append(2*r+r2)
        s1.append(abs((r-10)*(r2-10)+5))

plt.scatter(x1, y1, c=z1, s=s1, alpha=0.5, cmap=cm.hsv)

_=plt.show()
```

Теперь возьмем другой вид графика - quiver. Данный график показывает стрелками такую информацию, как, например, направление течения. В него переаются координаты начала стрелок, а также вектор для стрелки. units='width' означает, относительно чего измеряется размер стрелок (в данном случае относительно ширины, но можно задать, например, в дюймах или пикселях). Также можно задать как определется направление стрелки - как угол/длина или как координаты.<br>
Всё о quiver <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.quiver.html#matplotlib.pyplot.quiver">здесь</a>.

```python
x1=[]
y1=[]
z1=[]
s1=[]
for r in range(20):
    for r2 in range(20):
        x1.append(r)
        y1.append(r2)
        z1.append(2*r+r2)
        s1.append(abs((r-10)*(r2-10)+5))

plt.quiver(x1, y1, z1, s1, units='width')

_=plt.show()
```

Итак, выше представлен четырехмерный график. Теперь добавим стрелкам цвет, чтобы получить пятимерный график.

```python
x1=[]
y1=[]
z1=[]
s1=[]
c1=[]
for r in range(20):
    for r2 in range(20):
        x1.append(r)
        y1.append(r2)
        z1.append(2*r+r2)
        s1.append(abs((r-10)*(r2-10)+5))
        c1.append(abs(r-r2))

plt.quiver(x1, y1, z1, s1, c1, units='width', cmap=cm.hot)

_=plt.show()
```

Теперь добавим к графику точки, которые будут показывать еще две координаты наших данных. Итого - семь измерений.

```python
x1=[]
y1=[]
z1=[]
s1=[]
c1=[]
sc=[]
ss=[]
for r in range(20):
    for r2 in range(20):
        x1.append(r)
        y1.append(r2)
        z1.append(2*r+r2)
        s1.append(abs((r-10)*(r2-10)+5))
        c1.append(abs(r-r2))
        sc.append(abs(math.sin(r/4)))
        ss.append(abs(math.cos(r2/4))*40)

plt.scatter(x1, y1, c=sc, s=ss, alpha=0.5, cmap=cm.winter)
plt.quiver(x1, y1, z1, s1, c1, units='width', cmap=cm.jet)

_=plt.show()
```

Мы не будем сегодня касаться отображения двумерных проекций трехмерных графиков, но с их помощью можно еще сильнее повысить размерность отображаемых данных.<br>

```python
# Импортируем необходимые элементы управления.
from ipywidgets import interact, IntSlider, FloatSlider, IntRangeSlider
```

Данная библиотека может работать следующим образом. При помощи метода `interact` мы связываем некоторую функцию с элемемнтом управления. При изменении значения в элементе управления будет вызываться переданная функция.<br>
Заметим, что связывание может проводиться <a href="https://ipywidgets.readthedocs.io/en/stable/examples/Using%20Interact.html">различым образом</a>.<br>
`IntSlider` создает слайдер, позволяющий выбрать целое значение.

```python
def f(x):
    return x

def f2(power):
  x=[i/100 for i in range(200)]
  y=[i**power for i in x]
  plt.plot(x, y)
  _=plt.show()
```

```python
_=interact(f, x=IntSlider(min=1, max=30, value=5, step=1))
```

```python
_=interact(f2, power=IntSlider(min=1, max=30, value=5, step=1))
```

`FloatSlider` создает слайдер с числами с плавающей точкой, `IntRangeSlider` - слайдер для задания интервала с границами целого типа.<br>
Как Вы можете видеть, можно связывать несколько элементов управления. Главное - не забыть указать имена параметров для передачи в функцию.

```python
def f3(power, interval):
  x=[i/100 for i in range(interval[0]*100, interval[1]*100)]
  y=[i**power for i in x]
  plt.plot(x, y)
  _=plt.show()
    
_=interact(f3, 
           power=FloatSlider(min=1, max=30, value=5, step=0.1, description='Power of function:', readout_format='.1f'),
           interval=IntRangeSlider(value=[0, 12], min=0, max=10, step=1, description='Interval:'))
```

Мы можем использовать не только слайдеры, но и текстовые поля. `BoundedIntText`позволяет ввести целое число из интервала.

```python
from ipywidgets import BoundedIntText
```

```python
_=interact(f2, power=BoundedIntText(min=1, max=30, value=5, step=1, description='Power:'))
```

Теперь посмотрим как можно взаимодействовать с элементами управления без использования `interact`.

`IntProgress` позволяет отобразить прогресс вычислений. Для этого создадим переменную `prg` и отобразим ее с использованием функции `display`. Далее можно просто обращаться к ее свойствам. Например, `value` хранит текущее значение (для всех элементов управления).

```python
from ipywidgets import IntProgress
import time
```

```python
prg=IntProgress(value=7, min=0, max=10, step=1, description='Loading:')
prg2=IntProgress(value=7, min=0, max=10, step=1, description='Loading:')

display(prg2)
display(prg)

for j in range(11):
  prg2.value=j
  for i in range(11):
    prg.value=i
    time.sleep(0.1)
```

Аналогично, мы можем создать переменную типа `BoundedIntText` и использовать ее в коде ниже.

```python
bit=BoundedIntText(min=1, max=30, value=5, step=1, description='Power:')
display(bit)
```

```python
f2(int(bit.value))
```

Теперь посмотрим на элементы управления, позволяющие выбрать или отобразить одно из двух значений.

```python
from ipywidgets import Checkbox, Valid, RadioButtons, SelectionSlider
```

`Checkbox` представляет собой простой селектор. `RadioButton` позволяет выбрать один из нескольких вариантов. `SelectionSlider` позволяет отображать варианты в виде слайдера.

```python
chk=Checkbox(value=False, description='Needs Interval')
bit2=BoundedIntText(min=1, max=30, value=5, step=1, description='Power:')
rng=IntRangeSlider(value=[0, 12], min=0, max=10, step=1, description='Interval:')

display(chk, bit2, rng)
```

`Valid` отображает булевскую переменную, но с точки зрения ее корректности.

```python
display(Valid(value=chk.value,description='Consider interval?'))

if chk.value:
    f3(int(bit2.value), rng.value)
else:
    f2(int(bit2.value))
```

```python
btns=RadioButtons(options=['with interval', 'without interval'], description='How to calculate?')
display(btns)
```

```python
if btns.value=='with interval':
    f3(int(bit2.value), rng.value)
else:
    f2(int(bit2.value))
```

```python
btns2=SelectionSlider(options=['with interval', 'without interval'], value='with interval',description='Consider interval?')
display(btns2)
```

```python
if btns2.value=='with interval':
    f3(int(bit2.value), rng.value)
else:
    f2(int(bit2.value))
```

У библиотеки есть масса других возможностей. Например, она позволяет выбрать цвет или осуществить выбор при помощи кнопок.

```python
from ipywidgets import ColorPicker, ToggleButtons
```

```python
clr=ColorPicker(description='Pick a color', value='blue')
btns3=ToggleButtons(options=['- Solid', '-- Dashed', '-. Dash-Point'], description='Line Style:')
display(clr, btns3)
```

```python
def f4(power):
  x=[i/100 for i in range(200)]
  y=[i**power for i in x]
  plt.plot(x, y, color=clr.value, linestyle=btns3.value.split(' ')[0])
  _=plt.show()
    
def f5(power, interval):
  x=[i/100 for i in range(interval[0]*100, interval[1]*100)]
  y=[i**power for i in x]
  plt.plot(x, y, color=clr.value, linestyle=btns3.value.split(' ')[0])
  _=plt.show()
```

```python
if btns2.value=='with interval':
    f5(int(bit2.value), rng.value)
else:
    f4(int(bit2.value))
```
